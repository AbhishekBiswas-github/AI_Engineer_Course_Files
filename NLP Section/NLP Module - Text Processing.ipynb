# Lowercasing Text

```python
sentence = "Her cat's name is Luna"
```

```python
lower_sentence = sentence.lower()
```

```python
lower_sentence
```

```python
sentence_list = [
    "Could you pass me the TV remote ?",
    "It is IMPOSSIBLE to find this hotel",
    "Want to go for dinner on Tuesday"
]
```

```python
lowercase_list = [sentence.lower() for sentence in sentence_list]
print(lowercase_list)
```

# Removing stop words

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords 
```

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
sentence = "It's not often you find a soggy banana on the street."
sentence_no_stopwords = ' '.join([word for word in sentence.split() if word not in en_stopwords])
```

```python
sentence_no_stopwords
```

# Stemming

```python
from nltk import PorterStemmer, SnowballStemmer, LancasterStemmer
```

```python
learn_tokens = ['learned', 'learning', 'learn', 'learns', 'learner', 'learners']
likes_tokens = ['likes', 'better', 'worse']
```

```python
ps = PorterStemmer()
for t in learn_tokens:
    print(t, " -> ", ps.stem(t))
```

```python
ls = LancasterStemmer()
for t in learn_tokens:
    print(t, " -> ", ls.stem(t))
```

```python
ls = SnowballStemmer('english')
for t in learn_tokens:
    print(t, " -> ", ls.stem(t))
```
