```python
from nltk import ngrams
from nltk.tokenize import word_tokenize
import pandas as pd
import matplotlib.pyplot as plt
```

```python
para = "Artificial Intelligence (AI) is revolutionizing the way we interact with technology, enabling machines to perform tasks that once required human intelligence. From voice recognition and image analysis to decision-making and language translation, AI systems are becoming increasingly sophisticated and integrated into everyday life. At its core, AI leverages algorithms and vast datasets to learn patterns, make predictions, and adapt to new information. This transformative technology is driving innovation across industries—healthcare, finance, transportation, and education—by improving efficiency, accuracy, and personalization. As AI continues to evolve, it raises important questions about ethics, privacy, and the future of work, prompting society to carefully consider how best to harness its potential while safeguarding human values."
```

```python
para
```

```python
import re
para_non_special = re.sub(r'[^a-zA-Z0-9\s]','',para)
print(para_non_special)
```

```python
tokens = word_tokenize(para_non_special)
print(tokens)
```

```python
unigram = pd.Series(ngrams(tokens, 1)).value_counts()
print(unigram[:10])
```

```python
unigram[:10].sort_values().plot.barh(color='green', width=0.9, figsize=(8, 4))
```

# Practical Exercise

### Importing Libraries

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
import pandas as pd
```

### Importing the dataset

```python
data = pd.read_csv('tripadvisor_hotel_reviews.csv')
```

### Displaying the information of the dataset

```python
data.info()
```

### Displaying the first 5 rows of the dataset

```python
data.head()
```

### Converting the Reviews in lowercase

```python
data['Review_lowercase'] = data['Review'].str.lower()
data.head()
```

### Removing the stopwords except "not"

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove('not')
data['Review_no_stopwords'] = data['Review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
data.head()
```

### Removing punctuation marks and replace "*" with "start"

```python
data['Review_no_punkt'] = data['Review_no_stopwords'].apply(lambda x: re.sub(r'[^a-zA-Z0-9*\s]', '', x))
```

```python
data['Review_no_punkt'] = data['Review_no_punkt'].apply(lambda x: re.sub(r'[*]', 'star', x))
```

```python
data.head()
```

### Tokenizing the Reviews

```python
data['tokenize'] = data['Review_no_punkt'].apply(lambda x: word_tokenize(x))
```

```python
data.head()
```

### Stemming the review

```python
ps = PorterStemmer()
data['stemmed'] = data['tokenize'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

### Lemmetizing the Reviews

```python
lammetizer = WordNetLemmatizer()
data['lemmatized'] = data['tokenize'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
data.head()
```

### Combining the all the tokens of all rows

```python
tokens_clean = sum(data['lemmatized'], [])
```

### Creating n-grams to check the frequency of words or combination of words

```python
unigram = pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts()
print(unigram)
```

```python
bigram = pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts()
print(bigram)
```

```python
trigram = pd.Series(nltk.ngrams(tokens_clean, 3)).value_counts()
print(trigram)
```
