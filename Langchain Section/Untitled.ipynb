```python
%load_ext dotenv
%dotenv
```

```python
import os
from openai import OpenAI
```

```python
openai_api_key = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(api_key=openai_api_key)
```

```python
response = client.chat.completions.create(model='gpt-4o-mini', messages=[{'role': 'system',
                                                                                     'content': '''You are Marv, a chatbot that reluctantly answer 
                                                                                     questions with sarcastic responses'''},
                                                                                    {'role': 'user',
                                                                                    'content': '''I've recently adopted a dog. Could you suggest some 
                                                                                    dog names?'''}])
```

```python
response
```

```python
print(response.choices[0].message.content)
```

```python
response2 = client.chat.completions.create(model='gpt-5-nano', messages=[{'role': 'system',
                                                                                     'content': '''You are Marv, an intelligent AI tool that classify the text
                                                                                      sentiment as positive, negative or neutral '''},
                                                                                    {'role': 'user',
                                                                                    'content': '''I loved the movie "Thamma" which released 2 weeks back'''}])
```

```python
response2
```

```python
print(response2.usage.total_tokens)
```

```python
print(response2.choices[0].message.content)
```

```python
response2 = client.chat.completions.create(model='gpt-4', messages=[{'role': 'user', 'content':'''Explain black hole in depth'''}],
                                          max_tokens=250, 
                                          temperature=0,
                                          seed=365,
                                          stream=True)


```

```python
print(response2.choices[0].message.content)
```
