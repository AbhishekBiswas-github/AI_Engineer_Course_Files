```python
from pinecone import Pinecone, ServerlessSpec
import os
from dotenv import load_dotenv, find_dotenv
```

```python
load_dotenv(find_dotenv(), override=True)
```

```python
pc = Pinecone(api_key=os.environ.get('PINECONE_API_KEY'))
```

```python
pc.list_indexes()
```

```python
index_name = 'my-index'
dim = 3
matric = 'cosine'
```

```python
if index_name in [index.name for index in pc.list_indexes()]:
    pc.delete_index(index_name)
    print(f"{index_name} successfully deleted")
else:
    print(f"{index_name} not present in the list")
```

```python
pc.list_indexes()
```

```python
pc.create_index(
    name=index_name, 
    dimension=dim, 
    metric=matric, 
    spec= ServerlessSpec(
        cloud='aws', 
        region='us-east-1'
    )
)
```

```python
pc.list_indexes()
```

```python
pc.create_index(
    name='sample-index', 
    dimension=1536, 
    metric='euclidean', 
    spec= ServerlessSpec(
        cloud='aws', 
        region='us-east-1'
    )
)
```

```python
pc.list_indexes()
```

```python
index_name
```

```python
index = pc.Index(name=index_name)
```

```python
response = index.upsert([
    ('Dog',[4.,0.,1.]),
    ('Cat',[4.,0.,1.]),
    ('Chicken',[2.,2.,1.]),
    ('Mantis',[6.,2.,3.]),
    ('Elephant',[4.,0.,1.]),
])
```
