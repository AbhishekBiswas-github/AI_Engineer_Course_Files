# Tuples

```python
# Exercise 1
Cars = ("BMW", "Dodge", "Ford")

# Exercise 2
Cars = ("BMW", "Dodge", "Ford")
Cars[-2]

# Exercise 3
name, age = 'Peter,24'.split(",")
print(name)
print(age)

# Exercise 4
def rectangle_info(length, width):
    return (length*width, 2*(length+width))

area, perimeter = rectangle_info(2, 10)
print(area)
print(perimeter)
```
