# Another way of defining the function

```python
# Exercise 1
def exponentiation_exp_2(x):
    print("Raised to the power of 2:")
    result =  x ** 2
    return result

exponentiation_exp_2(4)
```

# Using a function in another function

```python
# Exception 1
def plus_five(x):
    return x  +5

def m_by_3(x):
    return plus_five(x) * 3

m_by_3(5)
```

# Combining Conditional Statements and functions

```python
# Exercise 1
def compare_the_two(x, y):
    if x > y:
        print("Greater")
    elif x < y:
        print("Less")
    else:
        print("Equal")

compare_the_two(5, 5)
```
