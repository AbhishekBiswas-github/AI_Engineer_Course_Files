#  Use Conditional Statements and Loops Together

```python
# Exercise 1
for num in list(range(1, 11)):
    print(num * 2)

# Exercise 2
for num in range(1, 31):
    if num % 2 == 1:
        print(num, end = ' ')
    else:
        print('Even', end = ' ')
print('\n')

# Exercise 3
n = [1,2,3,4,5,6]
for i in n:
    print(i * 10, end = ' ')
print('\n')
for i in range(len(n)):
    print(n[i] * 10, end = ' ')
print('\n')
```

#  Conditional Statements, Functions, and Loops

```python
# Exercise 1
nums = [1,35,12,24,31,51,70,100]
def count(nums):
    count_less_20 = 0
    index = 0
    while index < len(nums):
        if nums[index] < 20:
            count_less_20 += 1
        index += 1
    return count_less_20
print(count(nums))
```

# Iterating over Dictionaries

```python
# Exercise 1
prices = {
    "box_of_spaghetti" : 4,
    "lasagna"  : 5,
    "hamburger" : 2
   }
quantity = {
    "box_of_spaghetti" : 6,
    "lasagna"  : 10,
    "hamburger" : 0
    }

money_spent = 0

for key in prices.keys():
    if prices[key] >= 5:
        money_spent += prices[key] * quantity[key]
    else:
        money_spent = money_spent

print(money_spent)

# Exercise 2
money_spent = 0

for key in prices.keys():
    if prices[key] < 5:
        money_spent += prices[key] * quantity[key]
    else:
        money_spent = money_spent

print(money_spent)
```
