# IF Statement

```python
# Exercise 1
if 5 > 2:
    print("The condition has been satisfied")

# Exercise 2
x, y = (10, 25)
if x > 3 and y > 13:
    print("Both conditions are correct")
if(x <= 3 or y <= 13):
    print("At least one of the conditions is false")
```

# ELSE Statement

```python
# Exercise 1
x = 234
if x > 100: 
    print("A busy day")
else:
    print("A calm day")
```

# ELIF Statement

```python
# Exercise 1
x = 234
if x > 200:
    print("Big")
elif (x > 100 and x <= 200):
    print("Average")
else:
    print("Small")

# Exercise 2
x = -234
if x > 200:
    print("Big")
elif (x > 100 and x <= 200):
    print("Average")
elif(x > 0 and x <= 100):
    print("Small")
else:
    print("Negative")
```
